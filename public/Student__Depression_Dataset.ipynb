# **Student Depression Dataset**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('Student Depression Dataset.csv')
```

```python
df.head()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
sns.countplot(data = df, x = 'Gender')
ax = sns.countplot(data = df, x = "Gender")
ax.bar_label(ax.containers [0])
plt.title('GENDER DISTRIBUTION')
plt.show()
```

```python
gb=df.groupby('Profession').agg({'Academic Pressure':'mean','Work Pressure':'mean','CGPA':'mean','Depression':'mean'})
gb
```

```python
sns.heatmap(gb, annot= True)
plt.title('[Profession Impact on Students]')
plt.show()
```

```python
df_male = df[df['Gender'] != 'Female']
df_male.head()
```

```python
gb_male =df_male.groupby('Profession').agg({'Academic Pressure':'mean','Work Pressure':'mean','CGPA':'mean','Depression':'mean'})
gb
```

```python
sns.heatmap(gb_male, annot= True)
plt.title('[Profession Impact on males]')
plt.show()
```
