```python
import numpy as np
import pandas as pd
```

```python
movie = pd.read_csv('tmdb_5000_movies.csv')
credit = pd.read_csv('tmdb_5000_credits.csv')
```

```python
movie.head()
```

```python
credit = credit.rename(columns={'movie_id': 'id'})
credit.head()
```

```python
movie = movie.merge(credit, on='id', suffixes=('_movie', '_credit'))
movie.head(2)
```

```python
movie = movie.rename(columns={'original_title': 'title'})
movie.head(2)
```

```python
movie.info()
```

```python
movies = movie[['id', 'title', 'overview', 'genres', 'keywords', 'cast', 'crew']]
movies.head(2)
```

```python
movies.dropna(inplace=True)
```

```python
movies.isnull().sum()
```

```python
movies.duplicated().sum()
```

```python
 movie.iloc[0].genres
```

```python
import ast
def convert(obj):
    if isinstance(obj, str):  #if obj is a string
        L = []
        for i in ast.literal_eval(obj):
            L.append(i['name'])
        return L
    # If it's already a list, just return it
    elif isinstance(obj, list):
        return obj
    else:  # Handle other cases as needed
        return []  # Or raise an exception
```

```python
movies['genres'] = movies['genres'].apply(convert)
movies
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
movies
```

```python
def convert_5(obj):
    if isinstance(obj, str):  #if obj is a string
        L = []
        counter = 0
        for i in ast.literal_eval(obj):
            if counter != 5:
                L.append(i['name'])
                counter+=1
            else:
                break
        return L
    # If it's already a list, just return it
    elif isinstance(obj, list):
        return obj
    else:  # Handle other cases as needed
        return []  # Or raise an exception
```

```python
movies['cast'] = movies['cast'].apply(convert_5 )
movies
```

```python

def fetch_director(obj):
    if isinstance(obj, str):  #if obj is a string
        L = []
        for i in ast.literal_eval(obj):
            if i['job'] == 'Director':
                L.append(i['name'])
                break
        return L
    # If it's already a list, just return it
    elif isinstance(obj, list):
        return obj
    else:  # Handle other cases as needed
        return []  # Or raise an exception
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
movies
```

```python
 movies['overview'] = movies['overview'].apply(lambda x:x.split())
 movies
```

```python
movies['genres'] = movies['genres'].apply(lambda x:[i.replace(" ","") for i in x])
movies['keywords'] = movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
movies['cast'] = movies['cast'].apply(lambda x:[i.replace(" ","") for i in x])
movies['crew'] = movies['crew'].apply(lambda x:[i.replace(" ","") for i in x])
movies
```

```python
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
movies.head()
```

```python
final_df = movies[['id', 'title', 'tags']]
final_df
```

```python
 final_df['tags'] = final_df['tags'].apply(lambda x:" ".join(x))
 final_df.head()
```

```python
 final_df['tags'][4]
```

```python
final_df['tags'] = final_df['tags'].apply(lambda x:x.lower())
```

```python
final_df.head()
```

```python
import nltk
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))
    string=" ".join(y)
    return string
```

```python
ps.stem('loved')
```

```python
final_df['tags'] = final_df['tags'].apply(stem)
final_df
```

```python
from sklearn.feature_extraction.text import CountVectorizer
count_v = CountVectorizer( stop_words='english',max_features=5000)
vector_tag = count_v.fit_transform(final_df['tags']).toarray()
```

```python
vector_tag
```

```python
count_v.get_feature_names_out()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(vector_tag)
```

```python
similarity
```

```python
similarity[0] #the first movie is showing the similarity with other movies in the output array.
```

```python
def recommend(movie):
    movie_index = final_df[final_df['title'] == movie].index
    if len(movie_index) == 0:  # Check if the movie title is found
        print(f"Movie '{movie}' not found in the dataset.")
        return  # Exit the function if the movie is not found

    movie_index = movie_index[0]  # Access the index if the movie is found
    distances = similarity[movie_index]
    movies_list = sorted(list(enumerate(distances)), reverse=True, key=lambda x:x[1])[1:6]
    for i in movies_list:
        print(final_df.iloc[i[0]].title)
```

```python
recommend('Avatar')
```

```python
recommend('Batman Begins')
```

```python
recommend('Hobbit: An Unexpected Journey')
```

```python
recommend('Spider-Man: No Way Home')
```
